```python
June = pd.read_csv("../Resources/June-citibike-tripdata.csv")
June_df= pd.DataFrame(June)
June_df.head()
```

```python
July = pd.read_csv("../Resources/July-citibike-tripdata.csv")
July_df = pd.DataFrame(July)
July_df.head()
```

```python
August = pd.read_csv("../Resources/August-citibike-tripdata.csv")
August_df = pd.DataFrame(August)
August_df.head()
```

```python
September = pd.read_csv("../Resources/September-citibike-tripdata.csv")
September_df = pd.DataFrame(September)
September_df.head()
```

```python
October = pd.read_csv("../Resources/October-citibike-tripdata.csv")
October_df = pd.DataFrame(October)
October_df.head()
```

```python
November = pd.read_csv("../Resources/November-citibike-tripdata.csv")
November_df = pd.DataFrame(November)
November_df.head()
```

```python
December = pd.read_csv("../Resources/December-citibike-tripdata.csv")
December_df = pd.DataFrame(December)
December_df.head()
```

```python
df_all =  pd.concat([June_df, July_df, August_df, September_df, October_df, November_df, December_df], ignore_index=True)
df_all.tail()
```

```python
df_all['gender']
```

```python
df_all['Gender'] = df_all['gender'].replace([0,1,2],['unknown',"Male","Female"])
df_all = df_all.drop('gender',1)
df_all["distance"] = ''
df_all.tail()
```

```python
df_all.isnull().any() # Null values in column "Birth Year" detected
```

```python
df_all.to_csv('../Resources/June_to_Dec_citibikenyc_tripdata.csv', index_label = "TripID")
```
